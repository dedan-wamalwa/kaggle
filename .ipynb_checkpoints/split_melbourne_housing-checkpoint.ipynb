```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
y=melb_data.Price
melb_features=['Rooms','Bathroom','Landsize','Lattitude','Longtitude']
X=melb_data[melb_features]
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=1)
```

```python
# optimizing the model by finding optimal point btn underfitting and overfitting
def find_mae(max_nodes,X_train,X_val,y_train,y_val):
    model=DecisionTreeRegressor(max_leaf_nodes=max_nodes,random_state=1)
    model.fit(X_train,y_train)
    pred_values=model.predict(X_val)
    mae=mean_absolute_error(y_val,pred_values)
    return mae
```

```python
# finding mean abs. error with different no of max tree nodes
node_error={}
for nodes in maxnodes:
    my_mae=find_mae(nodes,X_train,X_test,y_train,y_test)
    node_error[nodes]=my_mae
    print(f'Max leaf nodes : {nodes} \t\t Mean Absolute Error : {my_mae}')
min_error=min(list(node_error.values()))
best_max_nodes=[x for x in node_error if node_error[x]==min_error][0]
print(f'The best fit max leaf nodes is {best_max_nodes}')
```

```python
# we can therefore train our final model using 500 max nodes
melb_model=DecisionTreeRegressor(max_leaf_nodes=500,random_state=1)
melb_model.fit(X,y)
```
