```python
import pandas as pd
melb_data= pd.read_csv('melb_data.csv')
melb_data.columns
```

```python
melb_data.describe()
```

```python
melb_data=melb_data.dropna(axis=0)
```

```python
melb_data.describe()
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
y=melb_data.Price
melb_features=['Rooms','Bathroom','Landsize','Lattitude','Longtitude']
X=melb_data[melb_features]
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=1)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
melb_model=DecisionTreeRegressor(random_state=1)
melb_model.fit(X_train,y_train)
```

```python
print(melb_model.predict(X_train.head()))
print(y_train.head())
```

```python
print(melb_model.predict(X_test.head(10)))
print(list(y_test.head(10)))
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_absolute_percentage_error
print(mean_absolute_error(y_test,melb_model.predict(X_test)))
print(mean_absolute_percentage_error(y_test,melb_model.predict(X_test)))
```
